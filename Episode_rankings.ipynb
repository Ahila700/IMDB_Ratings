```python
top_200 = sqldf("""SELECT primaryTitle, averageRating, numVotes, 
         FROM title t INNER JOIN ratings r ON t.tconst = r.tconst
         WHERE numVotes > 50000
         AND averageRating > 8
         AND (titleType = 'tvMiniSeries' OR titleType = 'tvSeries')
         ORDER BY averageRating desc
         LIMIT 200""")
```

```python
sqldf("""SELECT *
         FROM top_200
         WHERE genres NOT LIKE '%Documentary%'
         AND genres NOT LIKE '%News%'
         AND genres NOT LIKE '%History%'
         AND genres NOT LIKE '%Talk-Show%'
         AND genres NOT LIKE '%Reality-TV%'
         AND averageRating > 8.6""")
```
